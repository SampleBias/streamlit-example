```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Function to fetch historical Bitcoin data
def fetch_bitcoin_data():
    url = "https://api.coindesk.com/v1/bpi/historical/close.json?start=2024-02-20&end=2024-03-20"
    response = requests.get(url)
    data = response.json()['bpi']
    df = pd.DataFrame(data, index=[0]).T
    df.index = pd.to_datetime(df.index)
    df.rename(columns={0: 'close'}, inplace=True)
    return df

# RSI calculation function
def calculate_RSI(data, period=14):
    # ************* Insert RSI calculation code here *************
    # Example placeholder (remove this and insert your actual RSI logic)
    data['RSI'] = data['close'].rolling(period).mean()
    return data

# Fetch Bitcoin data
bitcoin_data = fetch_bitcoin_data()

# Calculate RSI (Make sure you've filled in the calculation logic)
bitcoin_data = calculate_RSI(bitcoin_data)

# Plot price and RSI
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(bitcoin_data['close'])
plt.title("Bitcoin Price Chart")
plt.ylabel('Price')

plt.subplot(2, 1, 2)
plt.plot(bitcoin_data['RSI'])
plt.axhline(30, color='red', linestyle='--')
plt.axhline(70, color='green', linestyle='--')
plt.ylabel('RSI')
plt.xlabel('Date')

plt.tight_layout()
plt.show()

# Buy/Sell signals
for index, row in bitcoin_data.iterrows():
    if row['RSI'] < 30:
        print(f"Potential Buy Signal at {index}")
    elif row['RSI'] > 70:
        print(f"Potential Sell Signal at {index}")
```
